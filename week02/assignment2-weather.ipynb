```python
# Importing Libraries
import matplotlib.pyplot as plt
import pandas
from datetime import datetime
```

```python
# Read in the csv using Pandas
weather_df = pandas.read_csv("weatherreadings1.csv")
weather_df #view the csv
```

```python
#View a list of the column names:
weather_df.columns
```

```python
#Extract the time and dry bulb temperature columns
weather_df[["dryBulbTemperature_Celsius", "reportEndDateTime"]] #The timedata is stored as a string in the format '%Y-%m-%dT%H:%M:%S%z'
```

```python
# Convert the time data to a format that can be used ofr the plot. 
time_of_measurement = []
for date in weather_df["reportEndDateTime"]:
    date = datetime.strptime(date, '%Y-%m-%dT%H:%M:%S%z') 
    time_of_measurement.append(date.strftime("%H:%M"))
print(time_of_measurement)

```

```python
fig, ax = plt.subplots()
ax.scatter(
    time_of_measurement,
    weather_df["dryBulbTemperature_Celsius"],
    label="Drybulb Temperature", color="blue"

)
ax.scatter(
    time_of_measurement,
    weather_df["wetBulbTemperature_Celsius"],
    label="Wetbulb Temperature", color="red"
)
ax.set_xticks(ticks = time_of_measurement[::4])
ax.set_xticklabels(time_of_measurement[::4], rotation=90)
ax.set_xlabel(f"Time of Measurement \n(Measurements taken between 02nd and 03rd October, 2024)")
ax.set_ylabel("Drybulb Temperature (°C)")
ax.set_title("Temperature vs. Time 2nd - 3rd Oct, 2024",
             fontdict = {'fontsize': 'large','fontweight' : "bold",}
             )
ax.legend()
```
